```python
# ADA embedding
import os
import requests

prompt = 'What is the meaning of life?'

response = requests.post(
    url = "https://ragnalysis.openai.azure.com/openai/deployments/ada_embedding/embeddings?api-version=2023-05-15", 
    headers = { "Content-Type": "application/json", "api-key": os.getenv('OPENAI_KEY') }, 
    json = { "input": prompt }
).json()['data'][0]['embedding']

print("Embedding length: ", len(response))
print("Embedding: ", response)
```

```python
# GPT models
import os
import requests

prompt = 'What is the meaning of life?'
model = 'gpt35_4k' # gpt35_4k   gpt35_16k   gpt4_1106

response = requests.post(
    url = f"https://ragnalysis.openai.azure.com/openai/deployments/{model}/chat/completions?api-version=2023-05-15", 
    headers = { "Content-Type": "application/json", "api-key": os.getenv('OPENAI_KEY') }, 
    json = { 
        "messages": [{
            "role": "user",
            "content": prompt
        }],
        "temperature": 0.7,
        "top_p": 0.95,
        "frequency_penalty": 0,
        "presence_penalty": 0,
        "max_tokens": 200,
        "stop": None
    }
).json()['choices'][0]['message']['content']

print("Response: ", response)
```
